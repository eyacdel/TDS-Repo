```python
#5 point summary of numerical attributes
#Five-Number Summary
#The five-number summary involves the calculation of 5 summary statistical quantities: namely:
#    - Median: The middle value in the sample, also called the 50th percentile or the 2nd quartile.
#    - 1st Quartile: The 25th percentile.
#    - 3rd Quartile: The 75th percentile.
#    - Minimum: The smallest observation in the sample.
#    - Maximum: The largest observation in the sample.

#Summary description of the dataset
dataSet.describe().T
```

```python
from numpy import percentile
#Percentiles and Quartiles are useful when we need to identify the outlier in our data. 
#Percentiles divide the whole population into 100 groups where as quartiles divide the population into 4 groups
#p = 25: First Quartile or Lower quartile (LQ)
#p = 50: second quartile or Median
#p = 75: Third Quartile or Upper quartile (UQ)
#loc vs. iloc functions in Pandas
#loc is label-based, which means that we have to specify the name of the rows and columns that we need to filter out.
#On the other hand, iloc is integer index-based. So here, we have to specify rows and columns by their integer index.
#calculate quartiles

datasubSet=dataSet.loc[:,['age','bmi','children','charges']]

# calculate smallest and largest values in the data subset.
data_min, data_max = datasubSet.min(), datasubSet.max()

# calculate quartiles
quartiles = percentile(datasubSet, [25, 50, 75])

#datasubSet
#data_min
#data_max
#quartiles

# print 5-number summary

print('Minimum:')
print('----------------------------')
print(data_min)
print('\n')
print('Lower quartile (LQ):')
print('----------------------------')
print(quartiles[0])
print('\n')
print('Median/Second quartile:')
print('----------------------------')
print(data_min)
print('\n')
print('Upper quartile (UQ):')
print('----------------------------')
print(data_min)
print('\n')
print('Maximum:')
print('----------------------------')
print(data_min)

```

#### Question 1:- It is believed that people who smoke have higher charges than those who don’t, perform an analysis with the given data to confirm or disprove this. (or in other words : Do charges of people who smoke differ significantly from the people who don't?)

Following are the Hypothesis adopted to confirm/disapprove this question/theory.

Null Hypothesis:
Charges of people who smoke differ significantly from people who don't

Alternative Hypothesis:
Charges of people who smoke does not differ significantly from people who don't


```python
dataSet.smoker.value_counts()
```

```python
#Option1: boxplot showing the distribution of charges of smokers and non-smokers
plt.figure(figsize=(12,9))
sns.boxplot(x='smoker',y='charges',data=dataSet)
plt.title("Overall distribution of charges comparing smokers and non-smokers")
```

```python
# Replacing string values to numbers. Need to run one by one
dataSet['sex'] = dataSet['sex'].apply({'male':0, 'female':1}.get)
dataSet['smoker'] = dataSet['smoker'].apply({'yes':1, 'no':0}.get)
dataSet['region'] = dataSet['region'].apply({'southwest':1, 'southeast':2, 'northwest':3, 'northeast':4}.get)
```

```python
dataSet.head()
```

```python
# Also checking the Correlation betweeen 'charges' and 'smoker' using joinplot
sns.jointplot(x=dataSet['smoker'],y=dataSet['charges'])
```

```python
smoker_ = dataSet[dataSet['smoker']==1]
non_smoker = dataSet[dataSet['smoker']==0]
```

```python
plt.title('Distribution of Medical Costs for Smokers Vs Non-Smokers')
ax = sns.kdeplot(smoker_['charges'], bw=10000, label='smoker', color='r')
ax = sns.kdeplot(non_smoker['charges'], bw=10000, label='non-smoker')
plt.show()
```

Based on analysis of the above distributions, it is clearly seen that the charges of people who smoke more significantly from people who don't. Therefore, it is not s not high enough to reject our null hypothesis

Based on analysis of the above distributions, it is clearly seen that the BMI of males is quite different from the BMI of the females. Therefore, we can conlude that the bmi of males differ significantly from bmi of females and it is enough to reject our null hypothesis
